```python
length = 5305
```

```python
# builds model 

NetInput = Input(shape=(length, 4, 1))
model = models.Sequential()

conv_A1 = layers.Conv2D(10, (4, 4), activation='relu', input_shape=(length, 4, 1))
conv_B1 = layers.Conv2D(10, (4, 4), activation='relu', input_shape=(length, 4, 1))
conv_B2 = layers.Conv2D(10, (20, 1), padding='same', activation='relu', 
                        input_shape=(length, 4, 1))
conv_C = layers.Conv2D(10, (4, 1), activation='relu', input_shape=(1, length, 4, 1))
flatten = layers.Flatten()
dense = layers.Dense(1, activation='relu')


x1 = conv_A1(NetInput)
x2 = conv_B2(conv_B1(NetInput))
sum_ = layers.Add()([x1, x2])
y = conv_C(sum_)
z = dense(flatten(y))


model = Model(inputs=[NetInput], outputs=z)
```

```python
model.compile(optimizer="Adam", loss="mse", metrics=["mae"
                                , tf.keras.metrics.RootMeanSquaredError()])
```
